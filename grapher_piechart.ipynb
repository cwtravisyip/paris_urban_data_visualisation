```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
heat = [73600,77430,127810,115460,210520,
        220020,306300,280770,241820,282720,
       493320,552270,512690,463300,797550,
       1059530,686120,582660,533380,595880]
total = [115890,123240,203190,180420,330810,
        333340,472550,411910,367500,445280,
         771210,859240,858040,731740,1309440,
        1597540,1043570,908960,868280,939970]
hot_water = [15390,16740,26560,23030,43710,
            41380,60590,48870,44570,57290,
             103180,119040,127150,98800,189690,
            204730,134510,121170,120900,128820]
```

```python
energy = {'heat':heat,'total':total,'hot_water':hot_water}
energy = pd.DataFrame(energy)
```

```python
energy.head()
```

```python
energy['portion'] = energy['heat']/energy['total']
energy['others'] = energy['total']-energy['heat']-energy['hot_water']
```

```python
energy
```

```python
total_heat = energy['heat'].sum()
total_hot_w = energy['hot_water'].sum()
total_others = energy['others'].sum()
```

```python
value = [total_heat,total_hot_w,total_others]
key = ['Heat','Water Heating','Others']
```

```python
#define the plot style
plt.style.use('ggplot')
plt.rcParams["font.family"] = "sans-serif"
plt.rcParams["figure.autolayout"] = False

colors = ['#F05682','#F7A1B9','#B8CA7D']

fig, ax = plt.subplots(figsize=(20,10 ))
ax = plt.pie(value, labels = key,colors = colors,textprops={'fontsize': 25},startangle = 90)
plt.show()

fig.savefig('./end_use.png',dpi = 300,transparent = True,bbox_inche = 'tight')
```
